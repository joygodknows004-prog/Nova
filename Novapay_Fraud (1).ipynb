# NovaPay Fraudulent Transaction Detection

# Day 1 — Project Foundation and Initial Setup

## 1. Company Background

NovaPay is a digital cross-border money transfer company headquartered in Toronto, Canada. It operates mainly in the United Kingdom, Canada and the United States.

The company allows customers to send, receive and hold different currencies across international borders.

## 2. Business Problem

NovaPay faces fraud risks such as identity theft, account takeover, unauthorised payments and transaction laundering.

The company currently uses static fraud rules and manual transaction reviews. These methods may not adapt quickly to new fraud patterns and may become difficult to manage as transaction volume increases.

The business problem is to identify fraudulent transactions more effectively while reducing unnecessary alerts for legitimate customers.

### Internship Project

**Project Duration:** 10 Days  
**Project Type:** Supervised Binary Classification  
**Company:** NovaPay  
**Target Variable:** `is_fraud`


## 3. Project Objective

The main objective of this project is to build a machine learning model that can distinguish fraudulent transactions from legitimate transactions.

The model will use information about each transaction, customer, device, location and transaction behaviour to identify suspicious patterns.

The project also aims to:

- improve fraud detection recall by at least 15% compared with the existing rules-based system;
- maintain acceptable precision to reduce false fraud alerts;
- provide explanations for fraud predictions;
- prepare the model for real-time transaction scoring.

## 4. Machine Learning Problem

This project is a supervised binary classification problem.

It is supervised because the dataset already contains confirmed fraud labels that the model can learn from.

It is binary classification because the model predicts one of two possible classes:

- `0` = legitimate transaction
- `1` = fraudulent transaction

The target variable is `is_fraud`.


## 5. Project Success Criteria

The project will be considered successful if the machine learning model can detect fraudulent transactions better than the existing rules-based system.

The main success target is to improve recall by at least 15% compared with the current baseline while maintaining acceptable precision.

The model will be evaluated using the following metrics:

- **Recall:** measures how many actual fraudulent transactions the model correctly identifies.
- **Precision:** measures how many transactions predicted as fraud are truly fraudulent.
- **F1-score:** provides a balance between precision and recall.
- **ROC-AUC:** measures how well the model separates fraudulent transactions from legitimate transactions.
- **Accuracy:** measures the overall percentage of correct predictions, but it will not be used alone because the classes are imbalanced.

## 6. Business Meaning of the Metrics

High recall is important because missed fraud can cause financial loss.

High precision is important because too many false fraud alerts can delay legitimate customers and increase the workload of fraud analysts.

Therefore, the best model should detect as much fraud as possible without creating too many unnecessary alerts.


## 7. Project Stakeholders

The main stakeholders in this project are:

### Fraud Analysts
Fraud analysts need the model to identify suspicious transactions accurately and reduce unnecessary manual reviews.

### Customers
Customers expect their money transfers to be safe, fast and not delayed by incorrect fraud alerts.

### Compliance and Regulatory Teams
These teams need fraud decisions to be transparent, explainable and auditable.

### Business Management
Management wants to reduce financial losses, chargebacks and compliance penalties while protecting customer trust.

### Data Science and Engineering Teams
These teams are responsible for building, testing, deploying and maintaining the fraud detection system.

## 8. Stakeholder Expectations

The fraud detection model should:

- detect a high proportion of fraudulent transactions;
- avoid flagging too many legitimate transactions;
- provide understandable reasons for fraud predictions;
- support real-time transaction scoring;
- remain reliable as fraud patterns change.

## 9. Project Scope

This project will cover the following activities:

- assess the quality and structure of the transaction dataset;
- clean and prepare the data;
- explore fraud patterns;
- create useful features;
- build and compare machine learning models;
- evaluate the models using recall, precision, F1-score and ROC-AUC;
- explain model predictions;
- prepare the model for real-time scoring;
- create a simple monitoring and retraining plan.

## 10. Project Assumptions

The project assumes that:

- the `is_fraud` column contains reliable fraud labels;
- the available transaction and customer features are relevant to fraud detection;
- the supplied dataset is suitable for model development;
- historical patterns can help predict future fraud;
- the model will support human review rather than replace it completely.

## 11. Project Risks

Possible risks include:

- missing or incorrect data;
- class imbalance between legitimate and fraudulent transactions;
- false positives affecting genuine customers;
- false negatives allowing fraud to go undetected;
- changes in fraud behaviour over time;
- the model learning patterns that may not work well on future data.

## 12. Project Constraints

The main constraints are:

- the project must be completed within 10 days;
- the model must support real-time transaction scoring;
- the system must provide clear explanations for fraud decisions;
- the model must improve recall by at least 15% compared with the rules-based baseline;
- acceptable precision must be maintained to limit false positives.

## 13. Expected Project Outcome
